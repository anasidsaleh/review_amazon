```python
cookie = {}
def Searchreviews(url):
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
r = Searchreviews("https://www.amazon.fr/Withings-Nokia-Body-composition-corporelle/product-reviews/B071XW4C5Q/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber=2")

soup = sp(r.text, 'html.parser')
```

#**LOAD THE REVIEW LIST**

```python
div = soup.find_all("div", {"data-hook":"review"})
```

####Load the title review

```python
title_list = []

for i in div:
  title= i.find("a",{'data-hook':'review-title'}).text.strip()
  title_list.append(title)

print(len(title_list))
print(title_list)
```

#### Load the name of the reviewer

```python
name_list = []
for i in div :
  names = i.find("span", {'class': 'a-profile-name'}).text
  name_list.append(names)

print(len(name_list))
print(name_list)
```

#### Load the rate review

```python
rate_list =[]
for i in div:
  rate = i.find("span",{'class':'a-icon-alt'}).text
  rate_list.append(rate[0:3])
#for i in range(len(rate_list)):
#  rate_list[i] = rate_list[i][0:3]

print(len(rate_list))
print(rate_list)

```

####Load the text review

```python
review_text_list =[]

for i in div:
  review_text= i.find("span",{'data-hook':'review-body'}).text.strip()
  review_text_list.append(review_text)

print(len(review_text_list))
print(review_text_list)
```

#**Converting the list to a dict and importing to a csv file**

####converting the list to a dict

```python
data =[{'Nom': name, 'Titre': title, 'Texte': text, 'Evaluation': rating} for name,title,text,rating in zip(name_list,title_list,review_text_list,rate_list)]
```

####importing to a csv

```python
pd.DataFrame.from_dict(data)
```
